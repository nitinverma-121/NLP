```python
import fasttext
import pandas as pd
import re
```

```python
data= pd.read_csv('ecommerce_dataset.csv', names= ['type', 'about'])
data.head()
```

```python
data= data.dropna()
```

```python
def cleaning(text):
  text= re.sub('[^\w\s\']', ' ', text)
  text= re.sub(' +', ' ', text)
  return text.lower().strip()
```

```python
f= """saf 'floral' framed painting wood 30 inch x 10 inch special effect uv print textured sao297 painting made up in synthetic frame with uv textured print which gives multi effects and attracts towards it

this is an special series of paintings which makes your wall very beautiful and gives a royal touch a perfect gift for your special ones """
```

```python
cleaning(f)
```

```python
data['about']= data['about'].apply(lambda x: cleaning(x))
```

```python
data.type.replace("Clothing & Accessories", "Clothing_Accessories", inplace=True)
```

```python
data['type'].value_counts()
```

```python
data['about']= "__label__" + data['type'] + " " + data['about']
```

```python
data.head()
```

```python
data.describe
```

```python
# we dont need the type column
data= data.drop(columns=['type'], axis= 0)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test= train_test_split(data, test_size= 0.2, random_state= 2936)
```

```python
train.shape
```

```python
test.shape
```

```python
train.to_csv("train.csv", header= None, index= None)
```

```python
test.to_csv("test.csv", header= None, index= None)
```

```python
model= fasttext.train_supervised("train.csv")
```

```python
model.predict("wintech assemble desktop pc cpu 500 gb sata hdd 4 gb ram intel c2d processor 3")
```

```python
model.predict("how to win friends and influence people is such a good book")
```

```python
model.test("test.csv")
```
